```python
import requests
import json
import random
from pprint import pprint

# Function to get random chest_pain_values and question_values for a given key
def get_random_values(user_key, num_values, result_dict):
    if user_key in result_dict:
        df_list = result_dict[user_key]
        
        # Check if the requested number of values is greater than the available entries
        num_values = min(num_values, len(df_list))
        
        # Randomly select 'num_values' entries
        random_entries = random.sample(df_list, num_values)

        return random_entries
    else:
        return None


# Ask the user for a key
user_key = input("Enter a key from result_dict: ")

# Ask the user for the number of random values to retrieve
num_values = int(input("Enter the number of random values to retrieve: "))

# Get random values for the specified key
random_values = get_random_values(user_key, num_values, result_dict)

# Display the result
if random_values is not None:
    print(f"Random values for key '{user_key}':")
    for i, entry in enumerate(random_values, start=1):
        print(f"{i}. Chest Pain: {entry['Chest Pain']}, Question: {entry['question']}")

    # Prepare the payload for the API request
    api_payload = {
        "all_symtoms": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values],
        "current": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values]
    }

    # Convert the payload to JSON
    api_payload_json = json.dumps(api_payload)

    # API request
    url = "http://192.168.0.122:8006/ekg/both_symtoms_prediction/?action=v1"
    headers = {'Content-Type': 'application/json'}
    response = requests.post(url, headers=headers, data=api_payload_json)

    # Print API response
    pprint(response.json())
    response_json = response.json()
else:
    print(f"The key '{user_key}' is not found in result_dict.")
```

```python
import requests
import json
import random
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np


# Function to get random chest_pain_values and question_values for a given key
def get_random_values(user_key, num_values, result_dict):
    if user_key in result_dict:
        df_list = result_dict[user_key]
        
        # Check if the requested number of values is greater than the available entries
        num_values = min(num_values, len(df_list))
        
        # Randomly select 'num_values' entries
        random_entries = random.sample(df_list, num_values)

        return random_entries
    else:
        return None
    
# List of DataFrame names
df_names = ['Aortic Dissection', 'AS_AI', 'CHF', 'Eso rupture', 'GI', 'infarct', 'Ischemia', 'MS', 'PE', 'Pericarditis', 'Pulmonary']

# Number of random tests to generate
num_tests = 2

# Generate and print random tests
for _ in range(num_tests):
    # Randomly select a key
    user_key = random.choice(df_names)

    # Randomly select the number of values (between 5 and 8)
    num_values = random.randint(5, 8)

    # Get random values for the specified key
    random_values = get_random_values(user_key, num_values, result_dict)

    # Display the result
    if random_values is not None:
        print(f"Random values for key '{user_key}':")
        for i, entry in enumerate(random_values, start=1):
            print(f"{i}. Chest Pain: {entry['Chest Pain']}, Question: {entry['question']}")

        # Prepare the payload for the API request
        api_payload = {
            "all_symtoms": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values],
            "current": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values]
        }

        # Convert the payload to JSON
        api_payload_json = json.dumps(api_payload)

        # API request
        url = "http://192.168.0.122:8006/ekg/both_symtoms_prediction/?action=v1"
        headers = {'Content-Type': 'application/json'}
        response = requests.post(url, headers=headers, data=api_payload_json)

        # Print API response
        # print(response.text)
        response_json = response.json()
        print("-" * 50)
    else:
        print(f"The key '{user_key}' is not found in result_dict.")

    # Extract data for 'beta' and 'live'
    beta_data = response_json['beta']['data']
    live_data = response_json['live']['data']
    beta_testsuggestion = response_json['beta']['testsuggestion']
    live_testsuggestion = response_json['live']['testsuggestion']

    # Function to create decorated histograms with different colors for each bar
    def create_colored_histogram(data, title, xlabel, ylabel, name_key, prob_key):
        plt.figure(figsize=(12, 8))
        names = [entry[name_key] for entry in data]
        probabilities = [entry[prob_key] for entry in data]

        # Assign different colors to each bar
        colors = plt.cm.viridis(np.linspace(0, 1, len(names)))
        
        bars = plt.bar(names, probabilities, color=colors)

        # Add grid lines
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        # Add labels and title
        plt.title(title, fontsize=18)
        plt.xlabel(xlabel, fontsize=14)
        plt.ylabel(ylabel, fontsize=14)

        # Add bar labels
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 4), ha='center', va='bottom')

        # Add legend
        plt.legend(['Probabilities'], loc='upper right', fontsize=12)

        # Rotate x-axis labels for better readability
        plt.xticks(rotation=45, ha='right', fontsize=12)
        plt.yticks(fontsize=12)

        # Add title
        plt.tight_layout()
        plt.show()

    # Create decorated histograms with different colors for diseases in 'beta' and 'live'
    create_colored_histogram(beta_data, 'Disease Probabilities in Beta', 'Diseases', 'Probabilities', 'diseases_name', 'probability')
    create_colored_histogram(live_data, 'Disease Probabilities in Live', 'Diseases', 'Probabilities', 'diseases_name', 'probability')

    # Create decorated histograms with different colors for tests in 'beta' and 'live'
    create_colored_histogram(beta_testsuggestion, 'Test Probabilities in Beta', 'Tests', 'Probabilities', 'test_name', 'probability')
    create_colored_histogram(live_testsuggestion, 'Test Probabilities in Live', 'Tests', 'Probabilities', 'test_name', 'probability')   
```

```python
import requests
import json
import random
import csv
import os
import matplotlib.pyplot as plt  

# Function to generate and save graph images
def generate_graph(data, test_num, version):
    plt.figure(figsize=(8, 6))
    diseases_names = [entry['diseases_name'] for entry in data]
    probabilities = [entry['probability'] for entry in data]
    plt.bar(diseases_names, probabilities)
    plt.xlabel('Disease Name')
    plt.ylabel('Probability')
    plt.title(f'Graph for Test {test_num} - Version: {version}')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    graph_path = f'images/graph_{test_num}_{version}.png'
    plt.savefig(graph_path)  # Save the graph image
    plt.close()
    return graph_path  # Return the path of the saved graph image

# Function to get random chest_pain_values and question_values for a given key
def get_random_values(user_key, num_values, result_dict):
    if user_key in result_dict:
        df_list = result_dict[user_key]
        num_values = min(num_values, len(df_list))
        random_entries = random.sample(df_list, num_values)
        return random_entries
    else:
        return None

# List of diseases
diseases = ['Aortic Dissection', 'AS/AI', 'CHF', 'Eso rupture', 'GI', 'infarct', 'Ischemia', 'MS', 'PE', 'Pericarditis', 'Pulmonary']

# Number of random tests to generate
num_tests = 200

# Create directories for saving CSV files if they don't exist
beta_dir = 'beta'
live_dir = 'live'
os.makedirs(beta_dir, exist_ok=True)
os.makedirs(live_dir, exist_ok=True)
os.makedirs('images', exist_ok=True)  # Create directory to save graph images

# CSV headers
csv_headers = ['diseasesname', 'symtoms', 'output', 'modeloutput', 'status', 'graph_image']  # Add 'graph_image' column

# Initialize lists to store data for beta and live tests
beta_csv_data_all = []
live_csv_data_all = []

# Generate and print random tests
for test_num in range(1, num_tests + 1):
    # Randomly select a key
    user_key = random.choice(diseases)

    # Randomly select the number of values (between 5 and 8)
    num_values = random.randint(5, 8)

    # Get random values for the specified key
    random_values = get_random_values(user_key, num_values, result_dict)

    # Prepare the payload for the API request
    api_payload = {
            "all_symtoms": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values],
            "current": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values]
        }

    # Convert the payload to JSON
    api_payload_json = json.dumps(api_payload)


    # API request
    url = "http://192.168.0.122:8006/ekg/both_symtoms_prediction/?action=v1"
    headers = {'Content-Type': 'application/json'}
    response = requests.post(url, headers=headers, data=api_payload_json)
    
    response_json = response.json()

    # Extract data for 'beta' and 'live'
    beta_data = response_json['beta']['data']
    live_data = response_json['live']['data']

    # Select the top two diseases from beta and live data
    beta_top_diseases = [entry['diseases_name'] for entry in beta_data[:2]]
    live_top_diseases = [entry['diseases_name'] for entry in live_data[:2]]

    # Determine status based on whether there's a match between beta and live top diseases
    status = 'pass' if user_key in beta_top_diseases or user_key in live_top_diseases else 'fail'

    # Generate and save graph image for beta version
    beta_graph_path = generate_graph(beta_data, test_num, version='beta')

    # Generate and save graph image for live version
    live_graph_path = generate_graph(live_data, test_num, version='live')

    # Prepare data for beta CSV
    beta_csv_data = [user_key, ', '.join([entry['Chest Pain'] for entry in random_values]), user_key, ', '.join(beta_top_diseases), status, beta_graph_path]  # Add path to graph image
    beta_csv_data_all.append(beta_csv_data)

    # Prepare data for live CSV
    live_csv_data = [user_key, ', '.join([entry['Chest Pain'] for entry in random_values]), user_key, ', '.join(live_top_diseases), status, live_graph_path]  # Add path to graph image
    live_csv_data_all.append(live_csv_data)

# Write data to CSV files for beta and live tests
beta_csv_file = os.path.join(beta_dir, 'beta_tests.csv')
live_csv_file = os.path.join(live_dir, 'live_tests.csv')

with open(beta_csv_file, 'w', newline='') as beta_csv:
    writer = csv.writer(beta_csv)
    writer.writerow(csv_headers)
    writer.writerows(beta_csv_data_all)

with open(live_csv_file, 'w', newline='') as live_csv:
    writer = csv.writer(live_csv)
    writer.writerow(csv_headers)
    writer.writerows(live_csv_data_all)

print("All CSV files generated and saved successfully!")
```
