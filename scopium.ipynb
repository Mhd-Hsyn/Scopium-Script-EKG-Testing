```python
import os
import pandas as pd

# Input and output folders
input_folders = ['update_previous_history', 'updated_chestpain', 'updated_physical_finding', 'updated_riskfactor']
output_folder = 'merged_folder'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate over CSV files in the first folder
for file_name in os.listdir(os.path.join(input_folders[0])):
    if file_name.endswith('.csv'):
        dfs = []

        # Read corresponding CSV file from each folder
        for folder in input_folders:
            file_path = os.path.join(folder, file_name)
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                dfs.append(df)

        # Concatenate DataFrames along rows
        merged_df = pd.concat(dfs, ignore_index=True)

        # Save the merged DataFrame to the output folder
        output_path = os.path.join(output_folder, file_name)
        merged_df.to_csv(output_path, index=False)

        print(f"Merged CSV file '{file_name}' has been created in '{output_folder}'.")

print("All CSV files have been merged.")
```

```python
import pandas as pd

# Load DataFrames
df1 = pd.read_csv('id_with_symptom/Aortic Dissection.csv')
df2 = pd.read_csv('id_with_symptom/AS_AI.csv')
df3 = pd.read_csv('id_with_symptom/CHF.csv')
df4 = pd.read_csv('id_with_symptom/Eso rupture.csv')
df5 = pd.read_csv('id_with_symptom/GI.csv')
df6 = pd.read_csv('id_with_symptom/infarct.csv')
df7 = pd.read_csv('id_with_symptom/Ischemia.csv')
df8 = pd.read_csv('id_with_symptom/MS.csv')
df9 = pd.read_csv('id_with_symptom/PE.csv')
df10 = pd.read_csv('id_with_symptom/Pericarditis.csv')
df11 = pd.read_csv('id_with_symptom/Pulmonary.csv')

# List of DataFrame names
df_names = ['Aortic Dissection', 'AS_AI', 'CHF', 'Eso rupture', 'GI', 'infarct', 'Ischemia', 'MS', 'PE', 'Pericarditis', 'Pulmonary']

# Create a dictionary to store the results
result_dict = {}

# Iterate over each DataFrame
for df_name, current_df in zip(df_names, [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]):
    # Extract 'Chest Pain' and 'question' columns
    chest_pain_values = current_df['Chest Pain'].tolist()
    question_values = current_df['question'].tolist()

    # Create a dictionary for the current DataFrame
    df_dict = {}
    for i in range(len(chest_pain_values)):
        df_dict[f'entry_{i+1}'] = {'Chest Pain': chest_pain_values[i], 'question': question_values[i]}

    # Add the dictionary to the main dictionary with the DataFrame name as key
    result_dict[df_name] = df_dict

# Print the resulting dictionary
print(result_dict)
```

```python
import random

# Function to get random chest_pain_values and question_values for a given key
def get_random_values(user_key, num_values, result_dict):
    if user_key in result_dict:
        df_list = result_dict[user_key]
        
        # Check if the requested number of values is greater than the available entries
        num_values = min(num_values, len(df_list))
        
        # Randomly select 'num_values' entries
        random_entries = random.sample(df_list, num_values)

        return random_entries
    else:
        return None
```

```python
import requests
import json
import random

# Function to get random chest_pain_values and question_values for a given key
def get_random_values(user_key, num_values, result_dict):
    if user_key in result_dict:
        df_list = result_dict[user_key]
        
        # Check if the requested number of values is greater than the available entries
        num_values = min(num_values, len(df_list))
        
        # Randomly select 'num_values' entries
        random_entries = random.sample(df_list, num_values)

        return random_entries
    else:
        return None


# Ask the user for a key
user_key = input("Enter a key from result_dict: ")

# Ask the user for the number of random values to retrieve
num_values = int(input("Enter the number of random values to retrieve: "))

# Get random values for the specified key
random_values = get_random_values(user_key, num_values, result_dict)

# Display the result
if random_values is not None:
    print(f"Random values for key '{user_key}':")
    for i, entry in enumerate(random_values, start=1):
        print(f"{i}. Chest Pain: {entry['Chest Pain']}, Question: {entry['question']}")

    # Prepare the payload for the API request
    api_payload = {
        "all_symtoms": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values],
        "current": [{"id": entry['question'], "symtoms": entry['Chest Pain']} for entry in random_values]
    }

    # Convert the payload to JSON
    api_payload_json = json.dumps(api_payload)

    # API request
    url = "http://192.168.0.122:8006/ekg/both_symtoms_prediction/?action=v1"
    headers = {'Content-Type': 'application/json'}
    response = requests.post(url, headers=headers, data=api_payload_json)

    # Print API response
    print(response.text)
else:
    print(f"The key '{user_key}' is not found in result_dict.")
```
